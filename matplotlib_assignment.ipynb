### We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below.
#### Task:
1. fitting it to the periodic function
2. plot the fit

#### Data

Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25

Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
months = np.arange(12)
plt.figure(figsize=(10,5))
plt.plot(months, max_temp, 'ro')
plt.plot(months, min_temp, 'bo')
plt.xlabel('Months')
plt.ylabel("Min and Max temperature")
plt.show()
```

```python
# 1. fitting it to the periodic function
from scipy import optimize
def year_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(year_temps, months, max_temp, [20,10,0])
res_min, cov_min = optimize.curve_fit(year_temps, months, min_temp, [-40,20,0])
```

```python
# 2. plot the fit
days = np.linspace(0, 12, num=365)
plt.figure(figsize=(10,5))
plt.plot(months, max_temp, 'ro')
plt.plot(days, year_temps(days, *res_max), 'r-')
plt.plot(months, min_temp, 'bo')
plt.plot(days, year_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius')
plt.show()
```

### This assignment is for visualization using matplotlib:
#### Data to use:  https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
df = pd.read_csv(url)
df.head()
```

#### 1. Create a pie chart presenting the male/female proportion

```python
gender = df['sex'].astype('category')
gender = gender.cat.codes
gender.head()
```

```python
plt.figure(figsize=(10,5))
plt.pie(gender.value_counts(), labels=df['sex'].unique(), autopct='%1.1f%%')
plt.title('Male/Female Proportion')
plt.show()
```

#### 2.Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
plt.figure(figsize=(10,5))
plt.scatter(df['age'], df['fare'],
            c = gender)
plt.xlabel("Age")
plt.ylabel("Fare Paid")
plt.show()
```
